```python
#Check whether any orders were placed before the customer's registration date.
```

```python
order_customer_check = pd.merge(
    orders,
    customers[["customer_id", "registration_date"]],
    on="customer_id",
    how="left"
)

invalid_order_dates = order_customer_check[
    order_customer_check["order_date"] < order_customer_check["registration_date"]
]

print("Orders placed before customer registration:", len(invalid_order_dates))
print("Percentage:", round(len(invalid_order_dates) / len(orders) * 100, 2), "%")
```

# 10,255 orders (20.51%) have order dates earlier than customer registration dates.
# These records are flagged as a data quality issue and retained because the correct source date cannot be determined.

### Observation

All date columns were successfully converted from `object` to `datetime64[ns]`. Monetary and numeric columns such as `total_amount`, `final_amount`, `gst_amount`, `amount`, and `return_amount` were already stored in appropriate numeric data types (`int64`/`float64`), so no additional conversion was required.

```python
#Check for outliers in final_amount — are there any orders above ₹5 lakh? Flag them.
```

```python
outliers = orders[orders["final_amount"] > 500000]

print("Number of Orders Above ₹5,00,000 :", len(outliers))

if len(outliers) > 0:
    print(outliers[["order_id", "customer_id", "final_amount"]])
else:
    print("No orders above ₹5,00,000 were found.")
```

```python
#Verify referential integrity — do all order_ids in order_items.csv exist in orders.csv?
```

```python
invalid_order_ids = order_items[~order_items["order_id"].isin(orders["order_id"])]

print("Invalid Order IDs :", len(invalid_order_ids))

if len(invalid_order_ids) > 0:
    print("\nRecords with Invalid Order IDs")
    print(invalid_order_ids["order_id"])
else:
    print("\nAll order_ids in order_items.csv exist in orders.csv.")
```

```python
#orders.csv: Check if delivered_date is always after order_date. Flag rows where it is not.
```

```python
invalid_deliverydates = orders[orders["delivered_date"] < orders["order_date"]]

print("invalid_deliverydates :", len(invalid_deliverydates))

if len(invalid_deliverydates) > 0:
    print("\nRecords with invalid_deliverydates")
    print(invalid_deliverydates[["delivered_date","order_date"]])
else:
    print("\ndelivered_date are always after order_date")
```

```python
#payments.csv: Count Failed vs. Success transactions. What is the failure rate?
```

```python
success = payments["status"].value_counts()["Success"]
failed = payments["status"].value_counts()["Failed"]
print("Success Transactions :", success)
print("Failed Transactions :", failed)

print("Failure Rate :", round((failed / len(payments)) * 100, 2), "%")
```

```python
#returns.csv: Does every return have a matching order_id in orders.csv?
```

```python
invalid_order_ids = returns[~returns["order_id"].isin(orders["order_id"])]

print("Invalid Order IDs :", len(invalid_order_ids))

if len(invalid_order_ids) > 0:
    print("\nRecords with Invalid Order IDs")
    print(invalid_order_ids[["order_id"]])
else:
    print("\nEvery return has a matching order_id in orders.csv.")
```

```python
#inventory.csv: How many products are in 'Out of Stock' status?
```

```python
out_of_stock = inventory["status"].value_counts()["Out of Stock"]
print(out_of_stock,"products are in out of stock status")
```
