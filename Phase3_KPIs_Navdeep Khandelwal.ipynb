```python
import pandas as pd
```

```python
orders = pd.read_csv("orders DA.csv")
order_items = pd.read_csv("order_items DA.csv")
customers = pd.read_csv("customers DA.csv")
products = pd.read_csv("products DA.csv")
payments = pd.read_csv("payments DA.csv")
returns = pd.read_csv("returns DA.csv")
inventory = pd.read_csv("inventory DA.csv")
suppliers = pd.read_csv("suppliers DA.csv")
```

```python
orders["order_date"] = pd.to_datetime(orders["order_date"], format="%d-%m-%Y")
orders["delivered_date"] = pd.to_datetime(orders["delivered_date"], format="%d-%m-%Y")

customers["registration_date"] = pd.to_datetime(customers["registration_date"], format="%d-%m-%Y")
customers["last_login_date"] = pd.to_datetime(customers["last_login_date"], format="%d-%m-%Y")

payments["payment_date"] = pd.to_datetime(payments["payment_date"], format="%d-%m-%Y")
payments["refund_date"] = pd.to_datetime(payments["refund_date"])

returns["return_date"] = pd.to_datetime(returns["return_date"], format="%d-%m-%Y")
returns["refund_date"] = pd.to_datetime(returns["refund_date"], format="%d-%m-%Y")

products["launch_date"] = pd.to_datetime(products["launch_date"], format="%d-%m-%Y")

inventory["last_restocked_date"] = pd.to_datetime(inventory["last_restocked_date"], format="%d-%m-%Y")

suppliers["created_date"] = pd.to_datetime(suppliers["created_date"], format="%d-%m-%Y")
```

```python
# KPI 1: Gross Merchandise Value (GMV)
```

```python
GMV = orders["final_amount"].sum()
print("Gross Merchandise Value (GMV): ₹", GMV)
```

```python
# KPI 2: Net Revenue
```

```python
Net_Revenue = orders[orders["status"] == "Delivered"]["final_amount"].sum()
print("Net Revenue: ₹", Net_Revenue)
```

```python
# KPI 3: Average Order Value (AOV)
```

```python
AOV = round(Net_Revenue / orders[orders["status"] == "Delivered"]["order_id"].count(),2)
print("Average Order Value (AOV): ₹", AOV)
```

```python
# KPI 4: Cancellation Rate
```

```python
Cancellation_Rate = round(orders[orders["status"] == "Cancelled"]["order_id"].count() / orders["order_id"].count() * 100, 2)
print(f"Cancellation Rate: {Cancellation_Rate}%")
```

```python
# KPI 5: Return Rate
```

```python
Return_Rate = round(returns["return_id"].count()/orders[orders["status"] == "Delivered"]["order_id"].count()*100,2)
print(f"Return Rate: {Return_Rate}%")
```

```python
# KPI 6: Customer Lifetime Value (CLV)
```

```python
CLV = round(customers["total_spent"].sum() / customers["customer_id"].count(),2)
print("Customer Lifetime Value (CLV): ₹", CLV)
```

```python
# KPI 7: Month-over-Month Growth
```

```python
Monthly_GMV = orders.groupby(orders["order_date"].dt.to_period("M"))["final_amount"].sum()

MoM_Growth = round((Monthly_GMV.iloc[-1] - Monthly_GMV.iloc[-2]) / Monthly_GMV.iloc[-2] * 100, 2)
print(f"Month-over-Month Growth: {MoM_Growth}%")
```

```python
# KPI 8: Top Category Revenue Share
```

```python
Category_Revenue = order_items.groupby(order_items["category"])["total_price"].sum()
Category_Revenue_sorted = Category_Revenue.sort_values(ascending= False).head(1)

Top_Category = Category_Revenue_sorted.index[0]
Top_Category_Revenue = Category_Revenue_sorted.iloc[0]

Top_Category_Revenue_Share = round(Top_Category_Revenue / order_items["total_price"].sum()*100,2)
print(f"Top Category: {Top_Category}")
print(f"Top Category Revenue Share: {Top_Category_Revenue_Share}%")
```

```python
# KPI 9:  Payment Failure Rate
```

```python
Payment_Failure_Rate = round(payments[payments["status"] == "Failed"]["payment_id"].count() / payments["payment_id"].count() * 100, 2)
print(f"Payment Failure Rate: {Payment_Failure_Rate}%")
```

```python
# KPI 10:  Inventory Fill Rate
```

```python
Inventory_Fill_Rate = round(inventory[inventory["status"] == "In Stock"]["inventory_id"].count() / inventory["inventory_id"].count() * 100, 2)
print(f"Inventory Fill Rate: {Inventory_Fill_Rate}%")
```
